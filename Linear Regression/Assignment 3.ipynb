### Task 1: Fish Dataset

This one's for warmup! Load in the Fish dataset, do some basic cleaning, and run linear regression to predict fish weight!

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
model = LinearRegression()
fish = pd.read_csv('Fish.csv')
fish
```

```python
one_hot = pd.DataFrame({'Species': list(set(fish['Species']))})
fish = pd.concat([fish,pd.get_dummies(fish['Species'], prefix='Species')],axis=1)
fish = fish.drop(columns = 'Species')
```

```python
weights = fish['Weight']
fishX = fish.drop(columns = 'Weight')
```

```python
fishModel = model.fit(X = fishX, y= weights)
fishModel.score(fishX, weights)
```

### Task 2: FIFA 19 Dataset

This one will require a little bit more artistry to get a good score. Load in the Fifa CSV and take a look at the contents. You'll have to perform some data cleaning and feature engineering on the data (removing columns, dealing with nulls, one hot encoding, etc). Once you've completed, run linear regression and see how your model does predicting overall score! Make sure you display any tables or graphs you decided to use!

```python
import numpy as np
from sklearn.model_selection import train_test_split
fifa = pd.read_csv('fifa.csv')
fifa = fifa.drop(columns = ['Name','Photo', 'Flag', 'Club Logo', 'ID', 'Nationality', 'Joined', 'Loaned From','Contract Valid Until', 'Jersey Number', 'Real Face','Unnamed: 0', 'Body Type', 'Work Rate','Value', 'Release Clause', 'Height'])
fifa
```

```python
positions = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
fifa = fifa.drop(columns = positions)
```

```python
fifa = pd.concat([fifa,pd.get_dummies(fifa['Club'], prefix='Club')],axis=1)
fifa = pd.concat([fifa,pd.get_dummies(fifa['Preferred Foot'], prefix='Preferred Foot')],axis=1)
fifa = pd.concat([fifa,pd.get_dummies(fifa['Position'], prefix='Position')],axis=1)
fifa = fifa.drop(columns = ['Club', 'Preferred Foot', 'Position'])
```

```python
a = fifa['Weight'].sort_values().fillna('0000').str[:-3].astype('int').replace(0, np.nan)
fifa['Weight'] = a.fillna(a.mean())
```

```python
fifa['Wage'] = fifa['Wage'].str[1:-1].replace(r'^\s*$', '000', regex=True).astype('int')
```

```python
fifa = fifa.dropna()
overallY = fifa['Overall']
fifa = fifa.drop(columns = 'Overall')
```

```python
X_train, X_test, y_train, y_test = train_test_split(fifa, overallY, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model = model.fit(X = X_train, y = y_train)
model.score(X_test, y_test)
```

```python
house = pd.read_csv('Housing.csv')
list(house.columns)
```

```python
new_house = house[['MSSubClass', 'MSZoning', 'LotArea', 'Neighborhood','BldgType', 'OverallQual', 'YearBuilt', 'YrSold','Foundation', 'FullBath',
 'HalfBath', 'BedroomAbvGr', 'GrLivArea', 'SaleCondition','BsmtFullBath', 'BsmtHalfBath','GarageArea', 'PoolArea']]
```

```python
def add_total_bathrooms(data):

    with_bathrooms = data.copy()
    bath_vars = ['BsmtFullBath', 'FullBath', 'BsmtHalfBath', 'HalfBath']
    weights = pd.Series([1, 1, 0.5, 0.5], index=bath_vars)
    with_bathrooms['TotalBathrooms'] = data[bath_vars].fillna(0).dot(weights)
    return with_bathrooms

new_house = add_total_bathrooms(new_house).drop(columns = ['BsmtFullBath', 'BsmtHalfBath','FullBath','HalfBath'])
```

```python
new_house = pd.concat([new_house,pd.get_dummies(new_house['MSZoning'], prefix='Zoning')],axis=1)
new_house = pd.concat([new_house,pd.get_dummies(new_house['Neighborhood'], prefix='Neighborhood')],axis=1)
new_house = pd.concat([new_house,pd.get_dummies(new_house['Foundation'], prefix='Foundation')],axis=1)
new_house = pd.concat([new_house,pd.get_dummies(new_house['MSSubClass'], prefix='Subclass')],axis=1)
new_house = pd.concat([new_house,pd.get_dummies(new_house['BldgType'], prefix='BldgType')],axis=1)
new_house = pd.concat([new_house,pd.get_dummies(new_house['SaleCondition'], prefix='SaleCondition')],axis=1)
new_house = new_house.drop(columns = ['MSZoning', 'Neighborhood', 'Foundation', 'MSSubClass', 'BldgType', "SaleCondition"])
```

```python
housey = house['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(new_house, housey, test_size=0.2, random_state=42)
model = LinearRegression()
model = model.fit(X = X_train, y = y_train)
model.score(X_test, y_test)
```
